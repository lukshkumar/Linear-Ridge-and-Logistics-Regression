```python
import pandas as pd
import numpy as np
from sklearn import feature_selection

df = pd.read_csv('AdvWorksCusts.csv')
result = pd.read_csv('AW_BikeBuyer.csv')

#Combining two dataframes from two different csv into one dataframe for training

print(df.shape)
print(result.shape)
df['BikeBuyer'] = result['BikeBuyer']
```

```python
#Finding the correlation of the numerical features with target variable.

#ID variables are always to be removed since they does not make any sense to use for training.

df.corr()
```

```python
# Analysing the categorical variables
important_categorical_varaibles = ['CountryRegionName','Education','Occupation','Gender','MaritalStatus']
important_categorical_varaibles = df[important_categorical_varaibles]
important_dummies = pd.get_dummies(important_categorical_varaibles)
target_variable = df['BikeBuyer']
important_dummies
```

```python
# Applying Chi Square to find the correlation between the two categorical variables.

feature_selection.chi2(important_dummies, target_variable)
```

```python
rfe_x  = important_dummies
rfe_x['NumberCarsOwned'] = df['NumberCarsOwned']
rfe_x['NumberChildrenAtHome'] = df['NumberChildrenAtHome']
rfe_x['TotalChildren'] = df['TotalChildren']
rfe_x['YearlyIncome'] = df['YearlyIncome']
rfe_x['HomeOwnerFlag'] = df['HomeOwnerFlag']
rfe_y  = df['BikeBuyer']

important_dummies.columns
```

```python
df = pd.read_csv('AdvWorksCusts.csv')
result = pd.read_csv('AW_AveMonthSpend.csv')

#Combining two dataframes from two different csv into one dataframe for training

print(df.shape)

print(result.shape)
df['AveMonthSpend'] = result['AveMonthSpend']
```

```python
# Analysing the categorical variables
important_categorical_varaibles = ['CountryRegionName','Education','Occupation','Gender','MaritalStatus']
important_categorical_varaibles = df[important_categorical_varaibles]
important_dummies = pd.get_dummies(important_categorical_varaibles)
```

```python
rfe_x  = important_dummies
rfe_x['NumberCarsOwned'] = df['NumberCarsOwned']
rfe_x['NumberChildrenAtHome'] = df['NumberChildrenAtHome']
rfe_x['TotalChildren'] = df['TotalChildren']
rfe_x['YearlyIncome'] = df['YearlyIncome']
rfe_x['HomeOwnerFlag'] = df['HomeOwnerFlag']
rfe_y  = df['AveMonthSpend']
important_dummies.columns
```

```python
# Recursive Feature Elimination

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

model = LinearRegression()

rfe = RFE(model)
rfe = rfe.fit(rfe_x,rfe_y)
# summarize the selection of the attributes
print(rfe.support_)
print(rfe.ranking_)

```

```python
X_data = important_dummies[['CountryRegionName_Australia', 'CountryRegionName_Canada',
       'CountryRegionName_France', 'CountryRegionName_Germany',
       'CountryRegionName_United Kingdom', 'CountryRegionName_United States', 'Occupation_Clerical',
       'Occupation_Management', 'Occupation_Manual', 'Occupation_Professional',
       'Occupation_Skilled Manual', 'Gender_F', 'Gender_M', 'MaritalStatus_M',
       'MaritalStatus_S']]
X_data['NumberCarsOwned'] = df['NumberCarsOwned']
X_data['NumberChildrenAtHome'] = df['NumberChildrenAtHome']
X_data['TotalChildren'] = df['TotalChildren']
X_data['YearlyIncome'] = df['YearlyIncome']
X_data['HomeOwnerFlag'] = df['HomeOwnerFlag']
Y_data  = df['AveMonthSpend']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_data, Y_data, test_size=0.3, random_state=0)

linearreg = LinearRegression()

linearreg.fit(X_train, y_train)
```

```python
pred = linearreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(linearreg.score(X_test, y_test)))
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, pred)
```
